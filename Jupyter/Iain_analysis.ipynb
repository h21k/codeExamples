## OK here is the calculations for the columns 
Note: I ended up not using pandas frames as numpy is way faster - and pandas blew up - at least on my machine :) I used: https://docs.scipy.org/doc/numpy-1.14.0/reference/generated/numpy.cov.html 

```python
# loop over transposed array columns then calculate covariance between
# 1st & 2nd
# 2nd & 3rd and so on - should be a total of 21 result arrays in total
# then store all in the sensor_covariances list
sensor_covariances = []
g = 0
h = 1
for column in data.T:
    print(column) # this just prints the transposed columns - just to check
    a = np.cov((data[:,g]),(data[:,h])) 
    while h < 22:
        sensor_covariances.append(a)
        g = g + 1
        h = h + 1
```
